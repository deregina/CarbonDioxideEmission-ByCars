### 2-5. Checking Outliers

```python
# check if quantitative variables have outliers.
```

```python
dtype_table = pd.DataFrame(data.dtypes, columns=['dtype'])
quantitative_variable_list = dtype_table.loc[dtype_table['dtype']!='object'].index.to_list()
quantitative_variable_list
```

Series.quantile : returns the value at the given quantile

```python
print('------------ Sorting out outliers using IQR (threshold 1.5 $*$ IQR) ------------')

for i in range(len(quantitative_variable_list)):
    Q1 = data[quantitative_variable_list[i]].quantile(0.25)
    Q3 = data[quantitative_variable_list[i]].quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5 * IQR

    data[quantitative_variable_list[i]+'_IQR_outlier'] = (data[quantitative_variable_list[i]] < (Q1 - threshold)) | \
                                                         (data[quantitative_variable_list[i]] > (Q3 + threshold))

    print(f'------------ quantitative column \'{quantitative_variable_list[i]}\' ------------')
    print(f'Q1: {Q1}  Q3: {Q3}  threshold: {threshold}\n')
    print(data[data[quantitative_variable_list[i]+'_IQR_outlier']][[quantitative_variable_list[i]]])
    num_outlier =len(data[data[quantitative_variable_list[i]+'_IQR_outlier']])

    print(f'\nnumber of outliers in column {quantitative_variable_list[i]} is {num_outlier}')
    print()
```

```python
print('Sorting out outliers using Z-score (threshold 3)')
threshold = 3

for i in range(len(quantitative_variable_list)):
    data[quantitative_variable_list[i]+'_Z'] = (data[quantitative_variable_list[i]] \
                                                - data[quantitative_variable_list[i]].mean())\
                                                / data[quantitative_variable_list[i]].std()
    data[quantitative_variable_list[i]+'_Z_outlier'] = abs(data[quantitative_variable_list[i]+'_Z'] > threshold)

    print(f'------------ quantitative column \'{quantitative_variable_list[i]}\' ------------')
    print(data[data[quantitative_variable_list[i]+'_Z_outlier']][[quantitative_variable_list[i], \
                                                                  quantitative_variable_list[i]+'_Z']])
    num_outlier =len(data[data[quantitative_variable_list[i]+'_Z_outlier']])

    print(f'\nnumber of outliers in column {quantitative_variable_list[i]} is {num_outlier}')
    print()
```
